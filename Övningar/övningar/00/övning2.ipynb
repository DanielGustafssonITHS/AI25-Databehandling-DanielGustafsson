```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the 'Totalt' sheet
df = pd.read_excel("komtopp50_2020.xlsx", sheet_name="Totalt", header=6, usecols="A:F")

df.columns = ["Rang 2020", "Rang 2019", "Kommun", "Folkmängd 2020", "Folkmängd 2019", "Förändring"]

# Quick exploration
print(df.head())
#print(df.info())
#print(df.describe())
```

```python
population_largest =  df.sort_values("Folkmängd 2020", ascending=False)
print(population_largest.head())
```

```python
population_smallest =  df.sort_values("Folkmängd 2020", ascending=True)
print(population_smallest.head())
```

```python
pop_2019 = df['Folkmängd 2019'].sum()
pop_2020 = df['Folkmängd 2020'].sum()
print(f"Folkmänden 2019 var {pop_2019}")
print(f"Folkmänden 2020 var {pop_2020}")
```

```python
pop_top_5 = df.nlargest(5, "Folkmängd 2020")
pop_min_5 = df.nsmallest(5, "Folkmängd 2020").sort_values(by="Folkmängd 2020", ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

pop_top_5.plot(kind='bar', x='Kommun', y='Folkmängd 2020', legend=False, ax=axes[0], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

axes[0].set_title('Sveriges 5 största kommuner 2020')
axes[0].set_xlabel('Kommun')
axes[0].set_ylabel('Folkmängd 2020')

pop_min_5.plot(kind='bar', x='Kommun', y='Folkmängd 2020', legend=False, ax=axes[1], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

axes[1].set_title('Sveriges 5 minsta kommuner 2020')
axes[1].set_xlabel('Kommun')
axes[1].set_ylabel('')

#plt.tight_layout()
plt.show()
```
