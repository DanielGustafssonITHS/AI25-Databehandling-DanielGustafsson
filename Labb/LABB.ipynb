### Korrelationsmatris fÃ¶r kostnader och prestationer

HÃ¤r undersÃ¶ks sambandet mellan variablerna `Personal`, `Expend`, `Outstate`, `Room.Board` och `Grad.Rate`.  

- `Personal` = uppskattade personliga utgifter per student.  
- `Expend` = skolans undervisningsutgifter per student.  
- `Outstate` = studieavgifter fÃ¶r studenter frÃ¥n andra delstater.  
- `Room.Board` = kostnader fÃ¶r boende och mat pÃ¥ campus.  
- `Grad.Rate` = examensgrad (andelen studenter som tar examen).  

Vi kan se att:
- `Expend` och `Grad.Rate` har ett svagt till mÃ¥ttligt positivt samband â€“ skolor som spenderar mer per student tenderar att ha nÃ¥got hÃ¶gre examensgrad.  
- `Outstate` korrelerar positivt med bÃ¥de `Expend` och `Personal`, vilket tyder pÃ¥ att skolor med hÃ¶gre avgifter ofta ocksÃ¥ har hÃ¶gre kostnader per student.

```python
# Korrelationsmatris mellan nÃ¥gra viktiga variabler
corr_columns = ['Personal', 'Expend', 'Outstate', 'Room.Board', 'Grad.Rate']
corr_matrix = college[corr_columns].corr()
corr_matrix
```

```python
# Visualisering av korrelationerna
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### Elite-status vs kostnader och prestationer

Boxplotsen visar skillnader mellan Elite-universitet (Yes) och icke-Elite (No):

- **Personal**: Elite-skolor har hÃ¶gre median och spridning, vilket visar att elever pÃ¥ dessa skolor generellt spenderar mer privat.  
- **Expend**: Elite-skolor spenderar mer pÃ¥ undervisning per student.  
- **Outstate**: Avgifter fÃ¶r utomstatliga studenter Ã¤r hÃ¶gre pÃ¥ Elite-skolor.  
- **Grad.Rate**: Elite-skolor har hÃ¶gre examensgrad, vilket antyder bÃ¤ttre studentprestationer.


```python
# JÃ¤mfÃ¶ra Elite vs numeriska variabler (alla i samma graf)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))  

# Personal spending
sns.boxplot(x='Elite', y='Personal', data=college, hue='Elite', dodge=False, ax=axes[0,0])
axes[0,0].set_title('Personal Spending by Elite Status')
axes[0,0].set_xlabel('Elite')
axes[0,0].set_ylabel('Personal Spending')


# Instructional expenditure 
sns.boxplot(x='Elite', y='Expend', data=college, hue='Elite', dodge=False, ax=axes[0,1])
axes[0,1].set_title('Instructional Expenditure by Elite Status')
axes[0,1].set_xlabel('Elite')
axes[0,1].set_ylabel('Expenditure')


# Out-of-state tuition 
sns.boxplot(x='Elite', y='Outstate', data=college, hue='Elite', dodge=False, ax=axes[1,0])
axes[1,0].set_title('Out-of-State Tuition by Elite Status')
axes[1,0].set_xlabel('Elite')
axes[1,0].set_ylabel('Out-of-State Tuition')


# Graduation rate 
sns.boxplot(x='Elite', y='Grad.Rate', data=college, hue='Elite', dodge=False, ax=axes[1,1])
axes[1,1].set_title('Graduation Rate by Elite Status')
axes[1,1].set_xlabel('Elite')
axes[1,1].set_ylabel('Graduation Rate')


plt.tight_layout()  
plt.show()
```

```python

sns.set_style('whitegrid')


fig, axes = plt.subplots(3, 2, figsize=(14, 18))

# Boxplots: Elite vs olika variabler

# Personal spending
sns.boxplot(x='Elite', y='Personal', data=college, hue='Elite', dodge=False, ax=axes[0,0])
axes[0,0].set_title('Personal Spending by Elite')

# Instructional expenditure
sns.boxplot(x='Elite', y='Expend', data=college, hue='Elite', dodge=False, ax=axes[0,1])
axes[0,1].set_title('Instructional Expenditure by Elite')

# Out-of-state tuition
sns.boxplot(x='Elite', y='Outstate', data=college, hue='Elite', dodge=False, ax=axes[1,0])
axes[1,0].set_title('Out-of-State Tuition by Elite')

# Graduation rate
sns.boxplot(x='Elite', y='Grad.Rate', data=college, hue='Elite', dodge=False, ax=axes[1,1])
axes[1,1].set_title('Graduation Rate by Elite')

# Alumni donations
sns.boxplot(x='Elite', y='perc.alumni', data=college, hue='Elite', dodge=False, ax=axes[2,0])
axes[2,0].set_title('Alumni Donations by Elite')

# Scatterplot: Applications vs Accepted

sns.scatterplot(x='Apps', y='Accept', hue='Elite', data=college, ax=axes[2,1])
axes[2,1].set_title('Applications vs Acceptances by Elite')

# Justera layout
plt.tight_layout()
plt.show()
```
